```python
data_dir = 'luad_tcga_pan_can_atlas_2018'
clinical_df = pd.read_csv(os.path.join(data_dir, 'data_clinical_patient.txt'),
                          sep='\t',
                          skiprows=4)
mrna_df = pd.read_csv(os.path.join(data_dir, 'data_mrna_seq_v2_rsem_zscores_ref_normal_samples.txt'),
                      sep='\t')
display(clinical_df.iloc[:10, :])
display(mrna_df.iloc[:10, :])
```

```python
## Step 2
clinical_df = clinical_df[['PATIENT_ID', 'OS_STATUS', 'OS_MONTHS']].copy()
print(f"결측치 처리 전: {clinical_df.shape[0]}")
clinical_df = clinical_df.dropna(subset=['OS_STATUS', 'OS_MONTHS'])
print(f"결측치 처리 후: {clinical_df.shape[0]}")
clinical_df = clinical_df[clinical_df['OS_MONTHS'] > 0]
print(f"OS_MONTHS 값 0 이하 제거 후: {clinical_df.shape[0]}")
clinical_df['OS_STATUS'] = clinical_df['OS_STATUS'].map({'0:LIVING': 0, '1:DECEASED': 1})
display(clinical_df.head())
```

```python
mrna_df = mrna_df.dropna(subset=['Hugo_Symbol'])

duplicates = mrna_df[mrna_df.duplicated(subset=['Hugo_Symbol'], keep=False)]
duplicates_sorted = duplicates.sort_values(by='Hugo_Symbol')
print(f"중복된 유전자 행 수: {len(duplicates_sorted)}개")
display(duplicates_sorted)

mrna_df = mrna_df.drop('Entrez_Gene_Id', axis=1, errors='ignore')
# groupby => 자동으로 Index로 설정
mrna_df = mrna_df.groupby('Hugo_Symbol').mean()
display(mrna_df.head())
```

```python
mrna_t = mrna_df.T
mrna_t['SAMPLE_ID'] = mrna_t.index
mrna_t['PATIENT_ID'] = mrna_t['SAMPLE_ID'].str[:12]
print(f"mrna_t 크기: {mrna_t.shape}")
display(mrna_t.head())

combined_df = pd.merge(clinical_df, mrna_t, on='PATIENT_ID', how='inner')
combined_df = combined_df.set_index('SAMPLE_ID')
combined_df = combined_df.drop('PATIENT_ID', axis=1)
print(f"최종 병합 데이터 크기: {combined_df.shape}")
display(combined_df.head())
```

```python
import pandas as pd
combined_df = pd.read_csv('/content/drive/MyDrive/tcga_luad_processed.csv', index_col=0)

from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(combined_df, test_size=0.3, random_state=42)
print(f"train_df: {train_df.shape}")
print(f"test_df: {test_df.shape}")

genes = combined_df.columns[2:].tolist()
print(f"combined_df[genes]: {len(genes)}")
```

```python
## Step 3
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multitest import fdrcorrection

# one-sample t-test(종양 샘플들의 평균이 정상 샘플 평균인 0과 다른지 검정)
dge_results = []
for gene in genes:
  tumor_expression = train_df[gene].dropna()
  if len(tumor_expression) >= 3:
    # t_stat은 사용하지 않는다
    t_stat, p_value = stats.ttest_1samp(tumor_expression, popmean=0.0)
    mean_z_score = tumor_expression.mean()
    dge_results.append({
        'Hugo_Symbol': gene,
        'mean_z_score': mean_z_score,
        'p_value': p_value
        })
dge_df = pd.DataFrame(dge_results).dropna()
dge_df['q_value'] = fdrcorrection(dge_df['p_value'])[1]

# significant gene의 비율이 20% 미만이 되도록 필터링
# 조건 1: q_value가 0.01 미만 (통계적으로 유의미함) - 조건1은 기여도가 낮다
# 조건 2: Z-score 평균의 절댓값이 2.58 초과 (발현량 차이가 확실함)
q_value_threshold = 0.01
z_score_threshold = 2.58
sig_genes_df = dge_df[
    (dge_df['q_value'] < q_value_threshold) &
    (dge_df['mean_z_score'].abs() > z_score_threshold)
]

print(f"combined_df[genes]: {len(genes)}")
print(f"sig_genes_df: {sig_genes_df.shape[0]}")
display(sig_genes_df.sort_values(by='q_value').head())

# 대표적인 종양 억제 유전자 TP53이 필터링을 통과했는지 확인
print('TP53' in sig_genes_df['Hugo_Symbol'].values)
display(dge_df[dge_df['Hugo_Symbol'] == 'TP53'])
```

```python
# mean_z_score를 mean_abs_z로 대체
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multitest import fdrcorrection

dge_results = []
for gene in genes:
  tumor_expression = train_df[gene].dropna()
  if len(tumor_expression) >= 3:
    # t_stat은 사용하지 않는다
    t_stat, p_value = stats.ttest_1samp(tumor_expression, popmean=0.0)
    mean_abs_z = tumor_expression.abs().mean()
    dge_results.append({
        'Hugo_Symbol': gene,
        'mean_abs_z': mean_abs_z,
        'p_value': p_value
        })
dge_df = pd.DataFrame(dge_results).dropna()
dge_df['q_value'] = fdrcorrection(dge_df['p_value'])[1]

# significant gene의 비율이 20% 미만이 되도록 z_abs_threshold를 상향
q_value_threshold = 0.01
z_abs_threshold = 2.81
sig_genes_df = dge_df[
    (dge_df['q_value'] < q_value_threshold) &
    (dge_df['mean_abs_z'] > z_abs_threshold)
]

print(f"combined_df[genes]: {len(genes)}")
print(f"sig_genes_df: {sig_genes_df.shape[0]}")
display(sig_genes_df.sort_values(by='q_value').head())

# 대표적인 종양 억제 유전자 TP53이 필터링을 통과했는지 확인
print('TP53' in sig_genes_df['Hugo_Symbol'].values)
display(dge_df[dge_df['Hugo_Symbol'] == 'TP53'])
```

```python
## Step 4
!pip install lifelines tqdm
from lifelines import CoxPHFitter
from tqdm.auto import tqdm
from statsmodels.stats.multitest import fdrcorrection
import pandas as pd
import numpy as np

survival_results = []
sig_genes = sig_genes_df['Hugo_Symbol'].tolist()

# 무결성 최종 점검
nan_counts = train_df[sig_genes+ ['OS_MONTHS', 'OS_STATUS']].isnull().sum().sum()
print(f"Non-missing: {nan_counts == 0}")

# 진행률 표시를 위한 tqdm()
for gene in tqdm(sig_genes):
  data_subset = train_df[['OS_MONTHS', 'OS_STATUS', gene]]
  try:
    cph = CoxPHFitter()
    cph.fit(data_subset, duration_col='OS_MONTHS', event_col='OS_STATUS')
    summary = cph.summary.loc[gene]
    survival_results.append({
      'Hugo_Symbol': gene,
      'HR': np.exp(summary['coef']),
      'p_value': summary['p']
    })
  except:
    continue
survival_df = pd.DataFrame(survival_results)
```

```python
# hr_threshold 설정
log_hrs = np.log(survival_df['HR'])
mu = log_hrs.mean()
sigma = log_hrs.std()
hr_threshold = np.exp(mu + 1 * sigma)
print(f"hr_threshold (Mean + 1SD): {hr_threshold:.4f}")
sig_survival_genes_df = survival_df[(survival_df['p_value'] < 0.05) & (survival_df['HR'] > hr_threshold)]

print(f"combined_df[genes]: {len(genes)}")
print(f"sig_survival_genes_df: {sig_survival_genes_df.shape[0]}")
display(sig_survival_genes_df.sort_values(by='HR', ascending=False).head())
```

```python
## Step 5
!pip install scikit-survival
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sksurv.linear_model import CoxnetSurvivalAnalysis

sig_survival_genes = sig_survival_genes_df['Hugo_Symbol'].tolist()

X_train = train_df[sig_survival_genes]
# scaler는 test_df에서도 train_df의 것을 적용
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)

y_train = np.array(
  list(zip(train_df['OS_STATUS'].astype(bool), train_df['OS_MONTHS'])),
  dtype=[('Status', '?'), ('Survival_in_months', '<f8')]
)

# Hyperparameter-1: l1_ratio=0.9 (L1 penalty : L2 penalty = 0.9 : 0.1)
coxnet = CoxnetSurvivalAnalysis(l1_ratio=0.9, fit_baseline_model=True)
coxnet.fit(X_train_scaled, y_train)

# Hyperparameter-2: target_count=15
target_count = 15

# l1_ratio=0.9 유지한 채로 alpha(penalty의 절대적 세기)를 바꿔감
coefficients = coxnet.coef_
non_zero_counts = (coefficients != 0).sum(axis=0)

# 후보군 개수가 15개와 가장 가까운 idx 선택
valid_indices = np.where((non_zero_counts >= 5) & (non_zero_counts <= 30))[0]
if len(valid_indices) > 0:
  best_idx = valid_indices[(np.abs(non_zero_counts[valid_indices] - target_count)).argmin()]
else:
  best_idx = (np.abs(non_zero_counts - target_count)).argmin()

selected_genes_mask = coefficients[:, best_idx] != 0
final_genes = X_train.columns[selected_genes_mask].tolist()
final_coefs = coefficients[selected_genes_mask, best_idx]

signature_df = pd.DataFrame({
  'Hugo_Symbol': final_genes,
  'Coefficient': final_coefs,
  'Multivariate_HR': np.exp(final_coefs)
}).sort_values(by='Coefficient', ascending=False)

print(f"signature_df: {len(signature_df)}")
display(signature_df)
```

```python
## Step 6
from sksurv.linear_model import CoxPHSurvivalAnalysis

X_train_final = X_train_scaled[final_genes]

# Model Training: L2 Penalty -> Multivariate Cox Regression
mv_cox_model = CoxPHSurvivalAnalysis(alpha=0.01)
mv_cox_model.fit(X_train_final, y_train)
model_coefs = pd.DataFrame({
  'Hugo_Symbol': final_genes,
  'Weight (Beta)': mv_cox_model.coef_,
  'Adjusted_HR': np.exp(mv_cox_model.coef_)
}).sort_values(by='Adjusted_HR', ascending=False)
display(model_coefs)

# median_risk: 환자를 고위험군, 저위험군으로 분할하는 중간값. test_df에서 사용.
train_df['Risk_Score'] = mv_cox_model.predict(X_train_final)
median_risk = train_df['Risk_Score'].median()
```

```python
## Step 7
import matplotlib.pyplot as plt
from sksurv.metrics import concordance_index_censored, cumulative_dynamic_auc
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test

X_test = test_df[sig_survival_genes]
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

X_test_final = X_test_scaled[final_genes]

y_test = np.array(
  list(zip(test_df['OS_STATUS'].astype(bool), test_df['OS_MONTHS'])),
  dtype=[('Status', '?'), ('Survival_in_months', '<f8')]
)

# 1. C-index: 0.5 무작위, 1.0 완벽
risk_scores_test = mv_cox_model.predict(X_test_final)
c_index = concordance_index_censored(y_test['Status'], y_test['Survival_in_months'], risk_scores_test)
print(f"Test Set C-index: {c_index[0]:.4f}")

# 2. Time-dependent ROC-AUC: AUC 0.5 무작위, 1.0 완벽
# 생존 기간 10~90 백분위수(percentile) 구간에 15개의 타임라인 설정
va_times = np.percentile(y_test['Survival_in_months'], np.linspace(10, 90, 15))
auc, mean_auc = cumulative_dynamic_auc(y_train, y_test, risk_scores_test, va_times)

plt.figure(figsize=(8, 5))
plt.plot(va_times, auc, marker="o", color='b', label='Time-dependent AUC')
plt.axhline(mean_auc, linestyle="--", color='r', label=f'Mean AUC = {mean_auc:.2f}')
plt.xlabel("Months after diagnosis")
plt.ylabel("AUC")
plt.title("Time-dependent ROC-AUC (Test Set)")
plt.legend()
plt.grid(True)
plt.show()

# 3-1. KM curve (High Risk vs Low Risk)
test_df['Risk_Score'] = risk_scores_test
test_df['Risk_Group'] = np.where(test_df['Risk_Score'] >= median_risk, 'High Risk', 'Low Risk')

# AUC 그래프와 x축 범위 통일
x_axis_limit = va_times.max()

kmf_high = KaplanMeierFitter()
kmf_low = KaplanMeierFitter()

plt.figure(figsize=(10, 6))

high_mask = (test_df['Risk_Group'] == 'High Risk')
kmf_high.fit(test_df[high_mask]['OS_MONTHS'],
             event_observed=test_df[high_mask]['OS_STATUS'],
             label='High Risk (Predicted)')
kmf_high.plot_survival_function(ax=plt.gca(), color='red')

low_mask = (test_df['Risk_Group'] == 'Low Risk')
kmf_low.fit(test_df[low_mask]['OS_MONTHS'],
            event_observed=test_df[low_mask]['OS_STATUS'],
            label='Low Risk (Predicted)')
kmf_low.plot_survival_function(ax=plt.gca(), color='blue')

# 3-2. Log-rank Test (High Risk vs Low Risk)
results = logrank_test(test_df[high_mask]['OS_MONTHS'],
                       test_df[low_mask]['OS_MONTHS'],
                       event_observed_A=test_df[high_mask]['OS_STATUS'],
                       event_observed_B=test_df[low_mask]['OS_STATUS'])
p_val_str = f"{results.p_value:.4e}" if results.p_value >= 0.0001 else "< 0.0001"

plt.xlim(0, x_axis_limit)
plt.title(f"Kaplan-Meier Survival Curves (Log-rank p-value: {p_val_str})")
plt.xlabel("Months")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True)
plt.show()
```
